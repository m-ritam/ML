```python
#Load the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing the dataset
dataset= pd.read_csv("/content/drive/My Drive/Excel file to use in colab/SocialNetworkAds.csv")
```

```python
dataset
```

```python
X=  dataset.iloc[:,[2,3]].values #Extracting only the independent variables
y= dataset.iloc[:,4].values #Extracting only the dependent variables
```

```python
X
```

```python
y
```

```python
#visualization by drawing a correration map
sns.heatmap(dataset.select_dtypes(include=['number']).corr(), annot=True)
```

```python
#splitting into Training and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc_X= StandardScaler()
X_train= sc_X.fit_transform(X_train)
X_test= sc_X.transform(X_test)
```

```python
# Fitting logistic regression to training dataset
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred= classifier.predict(X_test)
y_pred
```

```python
#visualizing the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set= X_train, y_train
X1, X2= np.meshgrid(np.arange(start= X_set[:,0].min()-1, stop= X_set[:,0].max()+1, step=0.01),
                    np.arange(start= X_set[:,1].min()-1, stop= X_set[:,1].max()+1, step=0.01))
plt.contourf(X1,X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('red','green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i,j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set==j, 0], X_set[y_set==j, 1],
              c=ListedColormap(('red','green'))(i), label=j)
plt.title('Logistic Regression (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated salary')
plt.legend()
plt.show()
```

```python
#Visualizing the Test Set results
from matplotlib.colors import ListedColormap
X_set, y_set= X_test, y_test
X1, X2 = np.meshgrid(np.arange(start=X_set[:,0].min()-1, stop=X_set[:,0].max()+1, step=0.01),
                     np.arange(start=X_set[:,1].min()-1, stop=X_set[:,1].max()+1, step=0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha=0.75, cmap=ListedColormap(('red','green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i,j in enumerate(np.unique(y_set)):
  plt.scatter(X_set[y_set==j, 0], X_set[y_set == j, 1],
              c = ListedColormap(('red','green'))(i), label=j)
plt.title('Logistic Regression (Test Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

```python
#confusion matrix evalution
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
cm
```
